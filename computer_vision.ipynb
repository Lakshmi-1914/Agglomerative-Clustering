```python
import cv2
from matplotlib import pyplot as plt
from google.colab import files
```

```python
image_file ='/content/Screenshot 2026-03-12 130300.png'
image = cv2.imread(image_file)
```

```python
image = cv2.imread(image_file)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
!wget -0 haar_face.xml https://github.com/opencv/opencv/raw/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade =cv2.CascadeClassifier("haar_face.xml")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors= 4)
```

```python
for(x,y,w,h) in faces:
  cv2.rectangle(image_rgb, (x,y), (x+w, y+h), (255,0,0),2)
```

```python
plt.figure(figsize=(15,8))
plt.imshow(image_rgb)
plt.title(f'Detected Faces : {len(faces)}')
plt.axis("off")
plt.show()
```
