```python
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("/content/Mall_Customers (1).csv")
print(df)
```

```python
X= df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
agg_model = AgglomerativeClustering(n_clusters=9)
```

```python
df['Cluster'] = agg_model.fit_predict(X)
```

```python
plt.scatter(X['Annual Income (k$)'],X['Spending Score (1-100)'],c=df['Cluster'],cmap='rainbow')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Agglomerative Clustering Example")
plt.show()
```
